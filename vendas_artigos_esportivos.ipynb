```python
from faker import Faker
import random
import pandas as pd

fake = Faker('pt_BR')

produtos = [
    {"produto_id": "TN023", "produto_nome": "Tênis de Corrida", "categoria": "Calçados"},
    {"produto_id": "CM456", "produto_nome": "Camiseta Dry Fit", "categoria": "Vestuário"},
    {"produto_id": "BM789", "produto_nome": "Bermuda Esportiva", "categoria": "Vestuário"},
    {"produto_id": "RL123", "produto_nome": "Relógio Esportivo", "categoria": "Acessórios"},
    {"produto_id": "GT456", "produto_nome": "Garrafa Térmica", "categoria": "Acessórios"},
    {"produto_id": "MC889", "produto_nome": "Meia de Compressão", "categoria": "Vestuário"},
    {"produto_id": "BA123", "produto_nome": "Bolsa de Academia", "categoria": "Acessórios"},
    {"produto_id": "CP456", "produto_nome": "Corda de Pular", "categoria": "Equipamentos"},
    {"produto_id": "FX458", "produto_nome": "Faixa de Pulso", "categoria": "Vestuário"},
    {"produto_id": "BE123", "produto_nome": "Boné Esportivo", "categoria": "Vestuário"}
]

canais = ["Varejo", "Oficina", "E-commerce"]
regioes = ["Sudeste", "Sul", "Nordeste", "Centro-Oeste", "Norte"]
cliente_tipos = ["Pessoa Física", "Pessoa Jurídica"]

def gerar_base_vendas(qtd_registros=1000):
    dados = []
    for _ in range(qtd_registros):
        sexo = random.choice(["Masculino", "Feminino"])
        nome = fake.first_name_male() if sexo == "Masculino" else fake.first_name_female()

        item = random.choice(produtos)
        produto = item["produto_nome"]
        categoria = item["categoria"]

        preco_custo = round(random.uniform(10, 500), 2)
        preco_unitario = round(random.uniform(30, 800), 2)
        quantidade = random.randint(1, 5)
        valor_total = round(preco_unitario * quantidade, 2)

        dados.append({
            "id_venda": fake.uuid4(),
            "data_venda": fake.date_between(start_date='-6M', end_date='today'),
            "nome": nome,
            "sexo": sexo,
            "cpf": fake.cpf(),
            "email": fake.email(),
            "telefone": fake.phone_number(),
            "produto": produto,
            "categoria": categoria,
            "quantidade": quantidade,
            "preco_custo": preco_custo,
            "preco_unitario": preco_unitario,
            "valor_total": valor_total,
            "canal_venda": random.choice(canais),
            "regiao": random.choice(regioes),
            "tipo_cliente": random.choice(cliente_tipos)
        })

    df = pd.DataFrame(dados)
    return df

# Gerar e salvar
df_vendas = gerar_base_vendas(500)
df_vendas.to_csv("base_vendas_esportivas.csv", index=False)

print("Base de vendas gerada com sucesso.")
```
